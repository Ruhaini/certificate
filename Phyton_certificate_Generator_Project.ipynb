```python
#in this function , the logic of creating a certificate will be inside!
#this function will receive 4 inputs (name,courseName,courseLevel,date)

#We are going to use this function 11 times

def certificate_generator(name,courseName,courseLevel,date):

  #store the generate file "pdf_file_name"
  #Creating a flie upfront for me to store my certificate with the detials
  pdf_file_name = "/content/drive/MyDrive/Colab Notebooks/Certificate generator/certificate files"+ name + "-" + courseName + "-" + courseLevel +".pdf"

  #Step 1 : I will prepare a blank sheet
  #Create a blanks canvas or a blank paper virtually using the function canvas.Canvas()
  #1st input ----> The storage path
  #2nd input ----> I need to specify what is the size and orientation of my canvas or my paper
  c = canvas.Canvas(pdf_file_name,pagesize = landscape(A4))

  #Step2 : Paste my template on my blank sheet of paper
  #I am going to use c.darwimage
  #1st input -----> I am retriving my template
  #2nd input -----> x-axis (Cordinate starts from 0)
  #3rd input -----> y-axis (COrdicate starts form 0 )
  #4th input -----> Width ---> A4[1](using a standard A4 variable that is imported)
  #5th input -----> height ---> A4[0](using a standard A4 variable that is imported)
  c.drawImage('/content/drive/MyDrive/Colab Notebooks/Certificate generator/Image20240405213727.jpg',0,0,A4[1],A4[0])

  # to populate name of participant-place it in the middle of the certificate/canvas (x-qaxis)
  #-place it in the middle of the a4 canvas (height y-axis)

  center_x = c._pagesize[0]/2

  center_y = c._pagesize[1]/2

  c.setFont("Lora-Bold",30)

  c.drawCentredString(center_x,center_y-50,name)

  c.drawCentredString(center_x,center_y-110,courseLevel)

  c.drawCentredString(center_x-190,center_y-160,date)

  c.setFont("Lora-Bold",30)

 #Populate the templae with unix time stamp (Certificate ID)
  cert_id = "Cert_ID:" + str(int(pd.Timestamp.now().timestamp()))
  c.setFont('Lora-Regular',12)
  c.drawCentredString(center_x + 266, center_y - 230,cert_id.upper())
  c.drawCentredString(center_x-190,center_y-160,date)


  c.save()
```
